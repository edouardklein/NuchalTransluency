# Chargement des données
## Ouverture des fichiers Excel

```python
import pandas as pd
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy
from collections import Counter
```

```python
data1 = pd.read_excel('Datas2013-1BIONUQUALOA1.xlsx')
data2 = pd.read_excel('Datas2013-2BIONUQUALOA1.xlsx')
data3 = pd.read_excel('Datas2014-1BIONUQUALOA1.xlsx')
```

```python
assert(all(data2.columns == data3.columns))
venn2([set(data1.columns), set(data2.columns)], set_labels=['Data2013-1', 'Data2013-2'])
print("Column in Data2013-1, but neither in Data2013-2 nor Data2014 : "+str(set(data1.columns) -set(data2.columns)))
print("Columns not in Data2013-1, but in Data2013-2 and Data2014 : "+str(set(data2.columns) -set(data1.columns)))
```

# CN et Age Maternel

## Extraction des données pertinentes

```python
data= pd.concat([data1, data2, data3])[['Date Nuque', 'DDN', 'MoM Nuque']].dropna()
data['Age Maternel'] = data['Date Nuque'] - data['DDN']
data['Age Maternel en années'] = data['Age Maternel'].dt.days/365
data[:10], len(data)
```

## Vérification de la cohérence des données

```python
sns.distplot(data['Age Maternel en années'])
```

```python
sns.distplot(data['MoM Nuque'])
```

## Corrélation

```python
plt.figure(figsize=(15,15))
with sns.axes_style("white"):
    sns.jointplot(x=data['Age Maternel en années'],
                  y=data['MoM Nuque'],
                  ylim=[0, 2],
                  kind="kde", color="k",
                  size=10,
                stat_func=scipy.stats.spearmanr)
```

Le même en excluant les enfants trisomiques des patientes agées.

```python
test_data = pd.concat([data2, data3])[['Date Nuque', 'DDN', 'MoM Nuque', ' T21']]
test_data[' T21'] = test_data[' T21'].fillna(0)
test_data = test_data.dropna()
test_data['Age Maternel'] = test_data['Date Nuque'] - test_data['DDN']
test_data['Age Maternel en années'] = test_data['Age Maternel'].dt.days/365
plt.figure(figsize=(15,15))
ix = (test_data[' T21'] != 'oui') | (test_data['Age Maternel en années'] < 38)
with sns.axes_style("white"):
    sns.jointplot(x=test_data['Age Maternel en années'][ix],
                  y=test_data['MoM Nuque'][ix],
                  ylim=[0, 2],
                  kind="kde", color="k",
                  size=10,
                stat_func=scipy.stats.spearmanr)
```

# CN et Age gestationnel
## Extraction des données pertinentes

```python
data= pd.concat([data1, data2, data3])[['Date Nuque', 'DDG', 'MoM Nuque']].dropna()
data['Age Gestationnel'] = data['Date Nuque'] - data['DDG']
data['Age Gestationnel en semaines'] = data['Age Gestationnel'].dt.days/7
data[:10], len(data)
```

## Vérification de la cohérence des données
Je n'arrive pas à faire une estimation de la distribution de l'age gestationnel.

```python
#sns.distplot(data['Age Gestationnel en semaines'])  # Long precessing, no resutls. Something's wrong Let's see what
ag = data['Age Gestationnel en semaines']
ag.max(), ag.min(), ag.mean(), ag.median()
```

Il s'avère qu'il y a des ages gestationnels négatifs ?!?

Et le max est à plus de 5000 semaines ?!?

Cherchons combien sont problématiques, fixons le max à 50 :

```python
ag = data['Age Gestationnel en semaines']
sum(ag<0), sum(ag>5000)
```

119 négatifs et 1 au dessus de 50, c'est pas énorme (on pourrait les virer et poursuivre tranquillement), mais c'est quand même pas top. voyons voir les dates.

```python
data[data['Age Gestationnel en semaines']<0]
```

A l'inspection visuelle, ce sont bien les dates qui sont foireuses (et non pas le calcul de la durée comme je le craignais).
Essayons de trouver des points communs.

```python
index_bad = data[data['Age Gestationnel en semaines']<0].index
counter = Counter(data1.loc[[i for i in index_bad if i in data1.index]]['Nom Labo'])
counter_df = pd.DataFrame.from_dict(counter, orient='index')
counter_df
counter_df.sort(inplace=True, columns=0)
counter_df.plot(kind='bar')
for lab in counter_df.index:
    print("Proportion of erroneous dates for lab"+str(lab))
    print(float(counter_df.loc[lab] / sum(data1["Nom Labo"] == lab)))
```

Bon, donc en gros on a quelques erreurs de saisies, dont une mère qui a recontré Doc Brown et est tombée enceinte en 2040, mais globalement le taux d'erreur reste ridiculement faible, sauf peut-être pour "HOPITAL MERE ENFANT" ou c'est suspérieur à 4%, il faudrait peut-être vérifier que tous leurs ordinateurs ont la date correctement réglée.

Finalement, je vais me contenter de laisser tomber ces données là et de passer à la suite.

```python
data = data[data['Age Gestationnel en semaines'] < 50]
data = data[data['Age Gestationnel en semaines'] > 0]
```
